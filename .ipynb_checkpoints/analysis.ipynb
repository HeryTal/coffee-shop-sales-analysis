```python
import pandas as pd

df = pd.read_excel(
    r"C:\Users\User\Desktop\Coffee shp\Coffee Shop Sales.xlsx"
)

```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

plt.hist(
    df['unit_price'],
    bins=20,
    edgecolor='white',
    alpha=0.85
)

plt.title("Distribution des prix", fontsize=16, fontweight='bold')
plt.xlabel("Prix (€)", fontsize=12)
plt.ylabel("Fréquence", fontsize=12)

plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.savefig("distributionDesPrix.png", dpi=300)
plt.show()


```

```python
import pandas as pd
import matplotlib.pyplot as plt

df['transaction_time'] = pd.to_datetime(
    df['transaction_time'],
    format='%H:%M:%S',
    errors='coerce'
)

df['hour'] = df['transaction_time'].dt.hour

df.groupby('hour')['unit_price'].sum().plot(kind='line', marker='o')

plt.title("Ventes par heure")
plt.xlabel("Heure")
plt.ylabel("Chiffre d'affaires")

plt.grid(alpha=0.5)
plt.savefig("VenteParHeure.png", dpi=300)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 7))

pivot = df.pivot_table(
    values='unit_price',
    index=df['transaction_time'].dt.hour,
    columns='product_category',
    aggfunc='sum'
)

sns.heatmap(
    pivot,
    cmap='YlOrRd',
    linewidths=0.5,
    linecolor='white',
    annot=False,
    cbar_kws={'label': 'Chiffre d’affaires'}
)

plt.title("Heatmap des ventes par heure et catégorie", fontsize=16, fontweight='bold')
plt.xlabel("Catégorie de produit", fontsize=12)
plt.ylabel("Heure de la journée", fontsize=12)

plt.tight_layout()
plt.savefig("heatmap.png", dpi=300)
plt.show()
```
